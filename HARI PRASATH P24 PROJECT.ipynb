```python
import tkinter as tk
from math import *

class Calculator:
    def __init__(self, master):
        self.master = master
        self.master.title("Scientific Calculator")
        self.master.geometry("400x500")
        self.master.resizable(width=False, height=False)

        self.result_var = tk.StringVar()

        self.entry = tk.Entry(self.master, textvariable=self.result_var, font=('Arial', 18), bd=10, insertwidth=4,
                              width=14, justify='right')
        self.entry.grid(row=0, column=0, columnspan=5)

        buttons = [
            '7', '8', '9', '/',
            '4', '5', '6', '*',
            '1', '2', '3', '-',
            '0', '.', '=', '+',
            'sin', 'cos', 'tan', 'sqrt',
            '(', ')', 'log', 'exp',
            'AC'
        ]

        row_val = 1
        col_val = 0

        for button in buttons:
            tk.Button(self.master, text=button, font=('Arial', 14), command=lambda b=button: self.button_click(b)).grid(
                row=row_val, column=col_val, ipadx=10, ipady=10)
            col_val += 1
            if col_val > 3:
                col_val = 0
                row_val += 1

    def button_click(self, button):
        current_text = self.result_var.get()

        if button == 'AC':
            self.result_var.set('')
        elif button == '=':
            try:
                result = str(eval(current_text))
                self.result_var.set(result)
            except Exception as e:
                self.result_var.set("Error")
        elif button in {'sin', 'cos', 'tan', 'sqrt', 'log', 'exp'}:
            try:
                result = str(eval(button + '(' + current_text + ')'))
                self.result_var.set(result)
            except Exception as e:
                self.result_var.set("Error")
        else:
            self.result_var.set(current_text + button)

if __name__ == "__main__":
    root = tk.Tk()
    calc = Calculator(root)
    root.mainloop()
```
